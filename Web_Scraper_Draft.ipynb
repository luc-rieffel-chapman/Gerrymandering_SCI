```python
#Import requests library to scrape data through URL
import requests
page = requests.get("https://www.politico.com/2020-election/results/wisconsin/")
page

#https://www.dataquest.io/blog/web-scraping-python-using-beautiful-soup/
```

```python
page.content
```

```python
#Beautiful soup is the library that we will use to parse data from the link
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#soup.prettify will translate the html content 
#we will now identify where in the html file the data is
print(soup.prettify())
```

```python
list(soup.children)
```

```python
soup.find_all(class_="outer-text")
#find class "outer text"
```
